```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import jieba

text= open(u'huge.txt','r',encoding="utf-8").read()

text=' '.join(jieba.cut(text))
print(text[:100])

wordcloud = WordCloud(background_color="white",font_path='simhei.ttf',width=800,height=660,margin=2).generate(text)


plt.imshow(wordcloud)
plt.axis("off")
plt.show()

wordcloud.to_file('test.png')
```
